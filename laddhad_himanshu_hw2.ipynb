# **Answer to Question 1:**
Instead of using a nested if statement inside a for loop, which gets tedious and cumbersome past the 4th level, a nested for loop with a if condition works for any number of levels with a shorter and more efficient code to build a tree structure same as that generated with the code used in class.

Logic -
1.   A "for" loop iterates all the links in the "links" list.
2. The first element in links is split based on "/" and stored in a "sublink_list". e.g. link="home/faculty" then sublink_list=['home','faculty'] and sub_link_length stores the length of the array used for iteration in the following "for" loop.
2.   Website_links is referenced as node in the second level for loop
3.   The second "for" loop goes through each element in the "sublink_list" and the if statement within it checks if the element is present in node as a key if not a empty dictionary as value is added with the element as the key
4.   The node is moved one level deeper and then step 3 is repeated.
5. Once teh secondary "for" loop ends, the next element in links array is loaded in the "link" variable and step 2 onwards is repeated till all the links are gone through.

```python
website_links = {}

for link in links:
    sublink_list = link.split("/")
    sub_link_length = len(sublink_list)

    node = website_links  #startiing from first level of the nested dict
    for j in range(sub_link_length):
        part = sublink_list[j]
        if part not in node:      # if this level doesn’t exist
            node[part] = {}       # make an empty dict
        node = node[part]         # move one step inside the dict
```
